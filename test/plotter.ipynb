```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

auc = pd.read_csv('./test/auc.csv')
disc = pd.read_csv('./test/disc.csv')
gen = pd.read_csv('./test/gen.csv')

auc = pd.read_csv('./FeGAN/auc.csv')
disc = pd.read_csv('./FeGAN/disc.csv')
gen = pd.read_csv('./FeGAN/gen.csv')

length = len(auc['values']) / 2
length = (int)(length)
idx = [*range(0, length)]
idx = idx + idx
auc["epochs"] = idx
disc["epochs"] = idx
gen["epochs"] = idx
```

```python
def plot_avg(disc, gen,auc,result_path):
    color = sns.color_palette("husl", 8)
    sns.set(style="darkgrid",palette=color)
    sns.lineplot(disc,x="epochs", y="values",linewidth = 2,label="Discriminator")
    sns.lineplot(gen,x="epochs", y="values",linewidth = 2,label="Generator")
    sns.lineplot(auc,x="epochs", y="values",linewidth = 2,label="AUC")
    #plt.savefig(result_path + "/" +"average"+".svg",format="svg",dpi=1200)
    
def plot(dy,gy,auc_y,k,seed,result_path):
    plt.style.use('ggplot')

    x = np.linspace(1, len(gy), len(gy))
    fig, ax = plt.subplots()
    ax.plot(x, gy["values"], color="cornflowerblue", label="Generator loss", linewidth=2)
    ax.plot(x, dy["values"],color="crimson", label="Average discriminator loss", linewidth=2)
    ax.plot(x, auc_y["values"], color="yellow", linewidth = 3, label="ROC AUC")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    ax.legend(loc="lower right")
    
```

```python
plot(disc,gen,auc,1,1,"")
```

```python
plot_avg(disc, gen, auc,"")
print(disc)
```
