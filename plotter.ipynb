```python
import matplotlib.pyplot as plt
import numpy as np

fst = array = np.linspace(0, 1, 200)
snd = np.logspace(np.log10(0.001), 0, 200)
```

```python
def plot(a,b):
    colors = ['#a1c9f4', '#8de5a1', '#ff9f9b', '#d0bbff', '#fffea3', '#b9f2f0']
    plt.style.use('ggplot')
    dy = a
    gy = b
    c = [1]-b
    
    x = np.linspace(1, len(gy), len(gy))
    fig, ax = plt.subplots()
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    ax.plot(x, gy, color="cornflowerblue", label="Generator loss",linewidth=2)
    ax.plot(x, dy, color="yellow", label="Average discriminator loss",linewidth=2)
    
    ax.plot(x, [1]-dy, color="pink", label="Average discriminator loss",linewidth=2)
    ax.plot(x, c, color="crimson", label="Average discriminator loss",linewidth=2)
    ax.legend(loc="upper left")
```

```python
plot(fst,snd)
```
